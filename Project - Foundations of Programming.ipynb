```python
# FIX CSV for many errors inside

# Posted variable should be a Date but for some strings there was 
# a State Name

# Questa funzione controlla se ci sono numeri in una stringa

def has_numbers(inputString): 
    return bool(re.search(r'\d', inputString))
id_errors = []
k = 0

# Se ci sono lettere nella variabile posted (dovrebbe essere una data) 
# l'id viene aggiunto in una lista

for i in dogs["posted"]:
    if has_numbers(dogs["posted"][k]) == True: 
        pass
    else:
        id_errors.append([dogs["id"][k]])
    k+=1
    
# Shift di variabili per le righe con problemi così che 
# in "posted" ci sia una data e non una città
    
for i in range(len(dogs["id"])):
  if dogs["id"][i] in id_errors:
    dogs.iloc[i:i+1,25:34] = dogs.iloc[i:i+1,25:34].shift(periods = 1, axis = 1)
  else:
    pass
```

**Query 12**

*12) Find for duplicates in the dogs dataset.*



```python
#group by similar dogs according to sex and primary and secondary breed
similardogs = dogs.groupby(["sex","breed_primary","breed_secondary"]).count() 

#select the dogs with same breeds and sex that appear more than once
similardogs = pd.DataFrame(similardogs[similardogs['id']>1]['id']) 

#reset indices
similardogs.reset_index(inplace = True)  
similardogs

# HTML(similardogs.to_html(index=False))
```

```python
df_sec_breed = []

# tramite un ciclo for racchiudi in una lista tutti i dataframe contenenti
# cani con stesso sesso, primary e secondary breed.

for x in range(similardogs.shape[0]): 
  df = dogs.loc[(dogs['sex']==similardogs.loc[x]['sex']) & (dogs['breed_primary']==similardogs.loc[x]['breed_primary'])&(dogs['breed_secondary']==similardogs.loc[x]['breed_secondary'])]
  df_sec_breed.append(df[["id","description"]])

#utilizzando sequencematcher su due stringe
#vengono utilizzate sequenze di stringe uguali e viene calcolato
#un punteggio di similarità.

from difflib import SequenceMatcher

# funzione per la similarità: ratio restituisce un punteggio di similarità
#tra 0 e 1 tra due stringhe.

# ratio è calcolato come: 2M/T; dove T è il numero totale di elementi ed M 
#è il numero totale di match.

def similar(a, b):
  if pd.isnull(a) or pd.isnull(b):
    return 0
  else:
    return SequenceMatcher(None, a, b).ratio()

from itertools import combinations



#per ogni dataframe contenuto in df_sec_breed viene calcolata la 
# somiglianza per ogni coppia possibile in quel dataframe.

for x in df_sec_breed:
  for index in list(combinations(x.index,2)):
    paired = pd.DataFrame(x.loc[index,:])
    if similar(paired.iloc[0]['description'], paired.iloc[1]['description']) >= 0.90:
      print(f"dog {paired.iloc[0]['id']} and dog {paired.iloc[1]['id']} are duplicates. Similarity is {similar(paired.iloc[0]['description'], paired.iloc[1]['description'])}")
    else:
      pass
print(x)
```
